Estamos criando uma collection de dados para gerar um Dataframe

```python
dados = {
'ID':[1,2,3,4],
'Nome': ['Alice','julia','lucas', 'daniel'],
'CPF': ['529.982.247-25','168.995.350-09','453.178.287-91','714.602.380-44'],
}
```

Transformando a collection em um dataframe

```python
dt = pd.DataFrame(dados)
```

Imprime na tela o dataframe

```python
dt.head()
```

Realizando tranformacao nos nomes das pessoas (UPPER)

```python
dt['Nome'] = dt['Nome'].str.upper()
```

```python
dt.head()
```

Filtrando os IDs 2,3 e 4 
<br>OBS: Sempre que for necessario realize o filtro dos dados, armazenar o resultado em uma variavel separada. Isso evita que os dodos originais se percam

```python
dt_filtrado = dt[(dt['ID']>=2) & (dt['ID'] <= 4) ]
```
