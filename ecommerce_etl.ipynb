```python
import pandas as pd
import mysql.connector
```

```python
connector = mysql.connector.connect(
    host='localhost',
    port=3307,
    user='root',
    password='senac',
    database='ecommerce_etl'
)
```

```python
query = """
select  c.nome, c.email, c.cidade,
        p.quantidade,p.data_pedido,
        pr.nome_produto, pr.preco
from clientes c
inner join pedidos p on c.id_cliente =p.id_cliente
inner join produtos pr on p.id_produto = p.id_produto
"""
```

```python
dt = pd.read_sql(query,connector)
dt.head()
```

Transformacao de dados e  processo de converter dados de um formato ou estrutura para outro,a fim de toma-los mais adequados para analise, visualizacao ou armazenamento. Isso pode envolver varias operacao ,como limpeza de dados, normalizacao,agregacao, filtragem e enriquecimento. A transformacao  de dados e uma etapa 

```python
dt['email'] = dt['email'].fillna('em branco')
dt.head()
```

```python
dt['quantidade'] = dt['quantidade'].fillna(1)
dt.head()
```

 verificacao do dataframe, para validar os tipos amazenamentos 

```python
dt.info()
```

```python
dt['data_pedido'] = pd.to_datetime(dt['data_pedido'])
dt.info()
```

adicionando data em linha que estava em branco

```python
dt['data_pedido'] = dt['data_pedido'].fillna(pd.Timestamp('2024-01-01'))
dt.head(40)
```

adicionando uma nova coluna com o nome"faturamento"

```python
dt['faturamento'] = dt['quantidade'] * dt['preco']
dt
```

```python
dt.to_excel('ecommerce_faturamento.xlsx', index=False)
```
